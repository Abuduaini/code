### 光谱数据多元高斯拟合  

```python
from scipy.optimize import curve_fit
```

```python
def fun3(x,a1,a2,a3,m1,m2,m3,s1,s2,s3):
    return (a1*np.exp(-((x-m1)/s1)**2)+
            a2*np.exp(-((x-m2)/s2)**2)+
            a3*np.exp(-((x-m3)/s3)**2))
```

```python
phone1_standard_popt,phone1_pcov = curve_fit(fun3,WL_interp,phone1_standard_spec,
                      bounds=([0,0,0,400,500,580,0,0,0],
                              [1,1,1,500,600,700,50,50,10]))

phone1_eyeprotect_popt,phone1_pcov = curve_fit(fun3,WL_interp,phone1_eyeprotect_spec,
                      bounds=([0,0,0,400,500,580,0,0,0],
                              [1,1,1,500,600,700,50,50,10]))

phone2_popt,phone2_pcov = curve_fit(fun3,WL_interp,phone2_standard_spec,
                      bounds=([0,0,0,400,500,580,0,0,0],
                              [1,1,1,500,600,700,20,50,50]))
phone2_eyeprotect_popt,phone2_eyeprotect_pcov = curve_fit(fun3,WL_interp,phone2_eyeprotect_spec,
                                                      bounds=([0,0,0,400,500,580,0,0,0],
                                                              [1,1,1,500,600,700,50,50,50]))
phone3_popt,phone3_pcov = curve_fit(fun3,WL_interp,phone3_standard_spec,
                      bounds=([0,0,0,400,500,580,0,0,0],
                              [1,1,1,500,600,700,50,50,50]))
phone3_eyeprotect_popt,phone3_eyeprotect_pcov = curve_fit(fun3,WL_interp,phone3_eyeprotect_spec,
                                                      bounds=([0,0,0,400,500,580,0,0,0],
                                                              [1,1,1,500,600,700,50,50,50]))
computer1_popt,computer1_pcov = curve_fit(fun3,WL_interp,computer1_spec,
                      bounds=([0,0,0,400,500,580,0,0,0],
                              [1,1,1,500,600,700,50,50,60]))

computer2_popt,computer2_pcov = curve_fit(fun3,WL_interp,computer2_spec,
                      bounds=([0,0,0,400,500,580,0,0,0],
                              [1,1,1,500,600,700,50,50,50]))
computer3_popt,computer3_pcov = curve_fit(fun3,WL_interp,computer3_spec,
                      bounds=([0,0,0,400,500,580,0,0,0],
                              [1,1,1,440,600,700,20,50,50]))
computer4_popt,computer4_pcov = curve_fit(fun3,WL_interp,computer4_spec,
                      bounds=([0,0,0,400,500,580,0,0,0],
                              [1,1,1,500,600,700,50,50,10]))
computer5_popt,computer5_pcov = curve_fit(fun3,WL_interp,computer5_spec,
                      bounds=([0,0,0,400,500,580,0,0,0],
                              [1,1,1,500,600,700,50,50,50]))
computer6_popt,computer6_pcov = curve_fit(fun3,WL_interp,computer6_spec,
                      bounds=([0,0,0,400,500,580,0,0,0],
                              [1,1,1,500,600,700,50,50,50]))
print('phone1_standard_popt')
print(phone1_standard_popt)
print('phone1_eyeprotect_popt')
print(phone1_eyeprotect_popt)
print('phone2_popt')
print(phone2_popt)
print('phone2_eyeprotect_popt')
print(phone2_eyeprotect_popt)
print('phone3_popt')
print(phone3_popt)
print('phone3_eyeprotect_popt')
print(phone3_eyeprotect_popt)
print('computer1_popt')
print(computer1_popt)
print('computer2_popt')
print(computer2_popt)
print('computer3_popt')
print(computer3_popt)
print('computer4_popt')
print(computer4_popt)
print('computer5_popt')
print(computer5_popt)
print('computer6_popt')
print(computer6_popt)
```

```python
def three_D_GAMUT_pointcloud(WL,popt,X_lambda,Y_lambda,Z_lambda):
    a=np.arange(0,256,15)
    a1=np.arange(1,10,2)
    a=np.append(a,values = a1)
    blue = popt[0]*np.exp(-((WL_interp-popt[3])/popt[6])**2)
    green = popt[1]*np.exp(-((WL_interp-popt[4])/popt[7])**2)
    red = popt[2]*np.exp(-((WL_interp-popt[5])/popt[8])**2)
    
    (blue_x,blue_y) =chroma_cordinate1931(blue,X_lambda,Y_lambda,Z_lambda)
    (blue_x1976,blue_y1976) = trans_1931_1976(blue_x,blue_y)
    (green_x,green_y) =chroma_cordinate1931(green,X_lambda,Y_lambda,Z_lambda)
    (green_x1976,green_y1976) = trans_1931_1976(green_x,green_y)
    (red_x,red_y) =chroma_cordinate1931(red,X_lambda,Y_lambda,Z_lambda)
    (red_x1976,red_y1976) = trans_1931_1976(red_x,red_y)
    
    point_cloud = np.zeros((3,3))
    point_cloud[0,:] = [blue_x1976,blue_y1976,0]
    point_cloud[1,:] = [green_x1976,green_y1976,0]
    point_cloud[2,:] = [red_x1976,red_y1976,0]
    
    for i in a:
        for j in a:
            for k in a:
                if (i==0 and j==0 and k==0):continue
                gauss = (blue*i+green*j+red*k)/255
                
                (gauss_x,gauss_y) =chroma_cordinate1931(gauss,X_lambda,Y_lambda,Z_lambda)
                (gauss_x1976,gauss_y1976) = trans_1931_1976(gauss_x,gauss_y)
                L = (popt[0]*i+popt[1]*j+popt[2]*k)/255
                
                xyl = np.zeros((1,3))
                xyl[0,:]=[gauss_x1976,gauss_y1976,L]
                point_cloud = np.append(point_cloud, values=xyl, axis=0)
    return(point_cloud)
```
